```python
# Netflix Dataset — Exploratory Data Analysis
# Analysis of 8,800+ titles on Netflix, exploring ratings, genres, release years, and more.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Data
df = pd.read_csv('../data/netflix_titles.csv')
df.head(10)
```

```python
# Data Overview & Cleaning
# Checking shape, data types, and missing values before analysis.
print("How many rows and columns does it have?\n", "Rows: ",df.shape[0],"\nColumns: ",df.shape[1],sep="",end="\n\n")
```

```python
print("What are the data types of each column?\n", df.dtypes,sep="",end="\n\n")
```

```python
missing_values_count = df.isnull().sum()
print("Which columns have missing values and how many?\n", missing_values_count,sep="",end="\n\n")
```

```python
# Distribution of Movies vs TV Shows
typesCount = df[:]["type"].value_counts()
plt.pie(typesCount.values, labels=typesCount.index, autopct='%1.1f%%')
plt.title("Distribution of Movies vs TV Shows")
plt.savefig('../visuals/movies_vs_tv_shows.png')
plt.close()

print("How many Movies vs TV Shows are there?\n", typesCount,sep="",end="\n\n")
```

```python
# Countries producing the most content
countriesMostContent = df[:]["country"].value_counts().head(10)
plt.bar(countriesMostContent.index, countriesMostContent.values)
plt.xlabel("Country")
plt.ylabel("Count")
plt.title("Top 10 Countries Producing the Most Content")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../visuals/top_countries_content.png')
plt.close()

print("Which countries produce the most content?\n", countriesMostContent,sep="",end="\n\n")
```

```python
# Most common ratings
mostCommonRatings = df["rating"].value_counts().head(10)
plt.bar(mostCommonRatings.index, mostCommonRatings.values)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Top 10 Most Common Ratings")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../visuals/most_common_ratings.png')
plt.close()

print("What are the most common ratings?\n", mostCommonRatings,sep="",end="\n\n")
```

```python
# Most common genres
topGenres = df[:]["listed_in"].str.split(",", expand=True).stack().str.strip().value_counts().head(10)
plt.bar(topGenres.index, topGenres.values)
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Top 10 Most Common Genres")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../visuals/top_genres.png')
plt.close()

print("What are the most common genres?\n", topGenres,sep="",end="\n\n")
```

```python
# Which release years have the most content added?
yearsMostContent = df[:]["release_year"].value_counts().head(10)
plt.bar(yearsMostContent.index, yearsMostContent.values)
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.title("Top 10 Release Years with the Most Content")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../visuals/top_release_years.png')
plt.close()

print("Which release years have the most content added?\n", yearsMostContent,sep="",end="\n\n")
```

```python
# Which months have the most content added?

# Define the correct calendar order
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Count occurrences of each month
monthsMostContent = df["date_added"].str.split(" ", expand=True)[0].value_counts()

# Reindex forces the months into calendar order
monthsMostContent = monthsMostContent.reindex(month_order)

plt.bar(monthsMostContent.index, monthsMostContent.values)
plt.xlabel("Month Added")
plt.ylabel("Count")
plt.title("Content Added Per Month (Jan → Dec)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../visuals/top_months_added.png')
plt.close()

print("Which months have the most content added?\n", monthsMostContent, sep="", end="\n\n")

print("Which months have the most content added?\n", monthsMostContent,sep="",end="\n\n")
```

```python
# What are the most common durations?
mostCommonDurations = df[:]["duration"].value_counts().head(10)
plt.bar(mostCommonDurations.index, mostCommonDurations.values)
plt.xlabel("Duration")
plt.ylabel("Count")
plt.title("Most Common Durations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../visuals/most_common_durations.png')
plt.close()

print("What are the most common durations?\n", mostCommonDurations,sep="",end="\n\n")
```
